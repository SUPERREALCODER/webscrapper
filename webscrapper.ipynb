```python
import requests
from bs4 import BeautifulSoup

```

```python

url ="https://www.kabaddiadda.com/league/1-pro-kabaddi-league/team/5012-gujarat-giants"
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
starting_element = soup.find("div", class_ ="tg-squad-list")
boxes = starting_element.find_all_next("div",class_ = "tf-squad-player")
print(boxes)
```

```python
import pandas as pd
from openpyxl import load_workbook
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE





url = input('enter')
html =  urlopen(url,context=ctx).read()
soup = BeautifulSoup(html,"html.parser")
div_tag = soup.find("div", class_="tg-squad-list")
tags = div_tag('a')
print(len(tags))
for tag in tags:

  tag_attributes = tag.attrs
  href_value = tag_attributes.get('href')
  split_url = href_value.split('/')

  desired_part = split_url[-1]
  print(desired_part)

```
